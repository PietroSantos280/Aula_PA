```python
def gerar_Senha():
  senha = ""
  for i in range(2):
    senha = senha + random.choice(string.ascii_uppercase) + random.choice(string.ascii_lowercase) + random.choice(string.digits) + random.choice(string.punctuation)
  return(senha)
```

##3) Crie uma função com retorno e com parâmetro chamada validar_senha(senha) que recebe uma string e retorna um dicionário contendo:

Chave_Significado

tamanho_valido -- Verdadeiro se tiver ≥ 8 caracteres

tem_numero -- Verdadeiro se possuir número

tem_maiuscula - Verdadeiro se tiver letra maiúscula

tem_minuscula -- Verdadeiro se tiver letra minúscula

tem_especial -- Verdadeiro se possuir caractere especial

senha_valida -- Verdadeiro apenas se todos os critérios forem atendidos

```python
def validar_senha(senha):
  tamanho_valido = len(senha)>=8
  tem_numero = any(c.isdigit() for c in senha)
  tem_maiuscula = any(c.isupper() for c in senha)
  tem_minuscula = any(c.islower() for c in senha)
  tem_especial = any(not c.isalnum() for c in senha)
  if   tamanho_valido and tem_numero and tem_maiuscula and tem_minuscula and tem_especial == True:
    return(print("Senha valida"))

  else:
    return(print("senha invalida"))
```

```python
validar_senha("Hv3>Wf2>")
```
