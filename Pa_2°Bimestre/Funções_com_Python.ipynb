# FUNÇÕES

Na programação, funções são blocos de código que realizam determinadas tarefas que normalmente precisam ser executadas diversas vezes dentro de uma aplicação. Quando surge essa necessidade, para que várias instruções não precisem ser repetidas, elas são agrupadas em uma função, à qual é dado um nome e que poderá ser chamada/executada em diferentes partes do programa.

```python
#exemplos de manipulação de retornos usando outras funções que foram realizadas anteriormente, retornando mais de um valor
def calculadora():
  calc = int(input("Calculadora Acionada, qual cálculo você quer fazer? \n Digite 1 para Somar \n Digite 2 para Subtrair \n Digite 3 para Multiplicar \n Digite 4 para Dividir"))
  valor1 = int(input("Digite o Primeiro Valor para fazer o Cálculo:"))
  valor2 = int(input("Digite o Segundo Valor para fazer o Cálculo:"))
  if calc == 1:
    resul = soma(valor1,valor2)
    tipo = " fez Soma"
  elif calc == 2:
    resul = sub(valor1,valor2)
    tipo =  "fez Subtração"
  elif calc ==3:
    resul = mult(valor1,valor2)
    tipo = "fez Multiplicação"
  elif calc ==4:
    resul = div(valor1,valor2)
    tipo = " fez Divisão"
  else:
    resul = "invalido"
    tipo = "invalido"
  if resul == "invalido":
    resul = "não digitou um valor válido pra fazer o cálculo"
  return resul, tipo
```

###2°Exercicio:
Conversor de Temperatura

Crie uma função chamada celsius_para_fahrenheit que recebe uma temperatura em Celsius ($C$) e retorna a temperatura equivalente em Fahrenheit ($F$).

$$
F = C \times \frac{9}{5} + 32
$$

 Lembre-se de converter a fração para decimal para garantir o cálculo correto e usar o return para devolver o valor.

```python
def celsius_para_fahrenheit(celsius):
    fahrenheit = celsius * 9 / 5 + 32
    return fahrenheit

t_celsius = 25
temperatura_f = celsius_para_fahrenheit(t_celsius)
print(f"{t_celsius}°C equivalem a {temperatura_f}°F")
```

Quadrado e Cubo

Crie uma função chamada calcular_potencias que recebe um número n e retorna o quadrado e o cubo desse número em uma tupla.

A função deve retornar dois valores. Lembre-se que em Python, você pode retornar múltiplos valores, que são automaticamente empacotados em uma tupla.

###6°Exercicio:
Cálculo de Desconto

Desenvolva uma função chamada calcular_desconto que aceita valor_original e percentual_desconto como parâmetros e retorna o valor final após aplicar o desconto.

$$
\text{Valor Final} = \text{Valor Original} \times (1 - \frac{\text{Percentual Desconto}}{100})
$$

 Para calcular o percentual corretamente, divida o percentual_desconto por 100 antes de subtrair de 1.

```python
def calcular_desconto(valor_original, percentual_desconto):
    valor_final = valor_original * (1 - percentual_desconto / 100)
    return valor_final
```

```python
preco = 200
desconto = 15
preco_final = calcular_desconto(preco, desconto)
print(f"O valor final após {desconto}% de desconto é R${preco_final:.2f}")
```

###7°Exercicio:
Maior e Menor

Crie uma função chamada encontrar_extremos que recebe uma lista de números e retorna o maior e o menor número da lista.

Você pode explorar as funções nativas do Python max() e min() para uma solução concisa, ou implementar a lógica de comparação manualmente

```python
def encontrar_extremos(numeros):
    maior = max(numeros)
    menor = min(numeros)
    return maior, menor

lista = [4, 7, 1, 9, 2, 3, 14]
maior_num, menor_num = encontrar_extremos(lista)
print(f"O maior número é {maior_num} e o menor número é {menor_num}.")
```

###8°Exercicio:
Calculadora Simples

Crie uma função chamada calculadora que recebe dois números (a e b) e uma operacao (string: 'soma', 'subtracao', 'multiplicacao', 'divisao') e retorna o resultado da operação.

 Use estruturas condicionais (if/elif/else) para determinar qual cálculo realizar e adicione uma verificação para evitar divisão por zero.

```python
def calculadora(a, b, operacao):
  if operacao == 'soma':
      return a + b
  elif operacao == 'subtracao':
      return a - b
  elif operacao == 'multiplicacao':
      return a * b
  elif operacao == 'divisao':
    if b != 0:
      return a / b
    else:
      return "erro:Divisão por zero!"
  else:
      return "operação inválida"
```

```python
print(calculadora(10, 5, 'soma'))
print(calculadora(10, 5, 'subtracao'))
print(calculadora(10, 5, 'multiplicacao'))
print(calculadora(10, 0, 'divisao'))
```

###10°Exemplo:
Sistema de Notas

Desenvolva uma função chamada avaliar_aluno que recebe uma nota (número) e retorna uma string indicando a situação do aluno: "Aprovado" (nota $\ge 7$), "Recuperação" ($\ge 5$ e $< 7$) ou "Reprovado" ($< 5$).

Lembre-se de usar operadores lógicos como and para definir os intervalos de notas de forma clara.

```python
def avaliar_aluno(nota):
    if nota >= 7:
        return "Aprovado"
    elif 5 <= nota < 7:
        return "Recuperação"
    else:
        return "Reprovado"

print(avaliar_aluno(8))
print(avaliar_aluno(6))
print(avaliar_aluno(3))
```
