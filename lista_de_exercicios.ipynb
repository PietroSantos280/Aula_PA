##calculo de média
Um fluxograma pode representar o processo de cálculo da média de duas notas, mostrando as etapas de entrada das notas, a operação de soma e a divisão por 2 para obter a média.

image.png

```python
# entrada
nota1 = float(input("Digite a nota da n1: "))
nota2 = float(input("Digite a nota da n2: "))
```

```python
# procassamento
media = (nota1 + nota2) / 2
```

```python
# saída
print(f"A média é: {media}")
```

##Conversão de moedas
Outro exemplo é a conversão de moeda. O fluxograma ilustra as etapas de receber a cotação do dólar, o valor em reais e a divisão para obter o valor em dólares.

image.png

```python
# entrada
cotacao = float(input("Digite a cotação do dólar: "))
valor_reais = float(input("Digite o valor em reais: "))
```

```python
# saida
print(f"O valor em dólares é: ${valor_dolar:.2f}")
```

##Cálculo de comissão
O cálculo de comissão pode ser facilmente representado por um fluxograma, mostrando as etapas de receber o preço unitário, a quantidade vendida e o cálculo da comissão.

image.png

```python
# entrada
preco_unitario = float(input("Digite o preço unitário da peça: "))
quantidade_vendida = int(input("Digite a quantidade vendida: "))

```

```python
# processamento
valor_total_venda = preco_unitario * quantidade_vendida
comissao = valor_total_venda * 0.05

```

```python
# saida
print(f"Valor total da venda: R${valor_total_venda:.2f}")
print(f"Comissão do vendedor: R${comissao:.2f}")
```

#EXERCICIO 3

Depois pergunte quanto ele ganha por hora e informe quanto ele irá ganhar. Sabendo que nessa empresa o funcionário ganha 70% a mais pelas horas trabalhadas.

#Exercicio 4

image.png

#Exercicio 5
Um grupo de amigos foi a um restaurante e deseja dividir a conta igualmente entre eles. Além disso, o restaurante cobra uma taxa de serviço de 10% sobre o valor total da conta mais uma gorjeta.

Escreva um algoritmo que solicite o valor da conta e o número de pessoas no grupo. Depois, calcule o valor total com a taxa de serviço e quanto cada pessoa deve pagar.

image.png

```python
#entrada
gorjeta =  float(input("quanto você deu de gorjeta para o garção: "))
conta =  float(input("quanto foi o valor final da conta: "))
qtpessoas = float(input("quantas pessoas estão com você: "))
```

```python
#processamento
taxadeservico = (conta*10)/100
valorfinal = (taxadeservico + conta + gorjeta)
pagoporpessoa = (valorfinal/qtpessoas)
```

```python
print(f"o valor final da sua conta foi de {valorfinal} reais cada pessoa deverá pagar cerca de {pagoporpessoa} reais")
```
